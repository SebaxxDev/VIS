al cargar los datos, se ignoran las advertencias ya que algunas celdas pueden tener formatos inconsistentes
provocando advertencias que no afectan el an√°lisis de los datos.

```python
# Celda 3: Carga Inteligente de Datos
import pandas as pd
import warnings
warnings.simplefilter("ignore")

# Funci√≥n para detectar d√≥nde empiezan los datos
def smart_read_excel(path, sheet_name):
    try:
        # Leemos primeras filas para escanear
        df_preview = pd.read_excel(path, sheet_name=sheet_name, header=None, nrows=5)
        target_row = 0
        
        # Buscamos fila con palabras clave del SIMCE
        for idx, row in df_preview.iterrows():
            row_str = row.astype(str).str.lower().tolist()
            # Si la fila tiene 'rbd' y 'prom', es la correcta
            if any('rbd' in s for s in row_str) and any('prom' in s for s in row_str):
                target_row = idx
                break
                
        print(f"[{sheet_name}] Cargando desde fila de encabezados: {target_row}")
        return pd.read_excel(path, sheet_name=sheet_name, header=target_row)
    except Exception as e:
        print(f"Error cargando {sheet_name}: {e}")
        return pd.DataFrame()

file_path = 'Datos/RBDSimce4toBasico2014-2024.xlsx'
sheets = ['2014', '2015', '2016', '2017', '2018', '2022', '2023', '2024']

# Carga masiva
dfs_dict = {}
print("Iniciando carga...")
for sheet in sheets:
    dfs_dict[sheet] = smart_read_excel(file_path, sheet)
print("Carga completada.")
```

Se unifica los DataFrames anuales en un √∫nico conjunto consolidado (`df_simce_all_years`). Dado que las columnas pueden variar entre a√±os (nuevas m√©tricas o cambios de metodolog√≠a), el c√≥digo realiza una **normalizaci√≥n de esquema** autom√°tica.

```python
# Celda 4: Procesamiento, Estandarizaci√≥n y Filtrado RM (Regi√≥n 13)
df_list_rm = []

# Mapeo Exacto basado en tus im√°genes
# Izquierda: Nombre Oficial | Derecha: Variaciones encontradas en los Excel
column_map = {
    'cod_reg_rbd': ['cod_reg_rbd', 'cod_region', 'region'],
    'cod_com_rbd': ['cod_com_rbd', 'cod_comuna', 'codigo_comuna'],
    'nom_com_rbd': ['nom_com_rbd', 'nom_comuna', 'nombre_comuna'],
    # Correcci√≥n clave para 2015/2017 seg√∫n tu foto:
    'prom_lect4b_rbd': ['prom_lect4b_rbd', 'prom_lect4b', 'prom_lect', 'ptje_lect4b_rbd'],
    'prom_mate4b_rbd': ['prom_mate4b_rbd', 'prom_mate4b', 'prom_mate', 'ptje_mate4b_rbd']
}

print("Procesando solo Regi√≥n Metropolitana (C√≥digo 13)...")

for year, df in dfs_dict.items():
    # 1. Normalizar encabezados
    df.columns = df.columns.str.lower().str.strip()
    
    # 2. Renombrar columnas clave
    for target, variations in column_map.items():
        if target not in df.columns:
            for v in variations:
                if v in df.columns:
                    df.rename(columns={v: target}, inplace=True)
                    break
    
    # 3. Validar columnas cr√≠ticas para filtrar
    if 'cod_reg_rbd' in df.columns and 'cod_com_rbd' in df.columns:
        
        # 4. Convertir Regi√≥n a N√∫mero para filtrar seguro
        # (A veces viene como "13" texto y otras como 13 n√∫mero)
        df['cod_reg_rbd'] = pd.to_numeric(df['cod_reg_rbd'], errors='coerce')
        
        # 5. FILTRO DURO: Solo Regi√≥n 13
        df_rm_year = df[df['cod_reg_rbd'] == 13].copy()
        
        if not df_rm_year.empty:
            # Aseguramos el a√±o
            if 'agno' not in df_rm_year.columns:
                df_rm_year['agno'] = int(year)
                
            # Convertimos puntajes a num√©rico (manejo de errores de tipeo en Excel)
            for col in ['prom_lect4b_rbd', 'prom_mate4b_rbd']:
                if col in df_rm_year.columns:
                    df_rm_year[col] = pd.to_numeric(df_rm_year[col], errors='coerce')
            
            # Seleccionamos solo columnas √∫tiles
            cols_to_keep = ['agno', 'cod_com_rbd', 'nom_com_rbd', 'prom_lect4b_rbd', 'prom_mate4b_rbd']
            # Filtramos solo las que existan
            cols_final = [c for c in cols_to_keep if c in df_rm_year.columns]
            
            df_list_rm.append(df_rm_year[cols_final])
            print(f"A√±o {year}: {len(df_rm_year)} registros de RM encontrados.")
        else:
            print(f"‚ö†Ô∏è A√±o {year}: No se encontraron datos para la Regi√≥n 13.")
    else:
        print(f"‚ùå A√±o {year} DESCARTADO: Falta columna 'cod_reg_rbd' o 'cod_com_rbd'. Columnas: {list(df.columns)}")

# Uni√≥n Final
if df_list_rm:
    df_simce_rm_raw = pd.concat(df_list_rm, ignore_index=True)
    print(f"\nTotal consolidado RM: {len(df_simce_rm_raw)} registros.")
else:
    raise ValueError("Error Cr√≠tico: No se carg√≥ ning√∫n dato para la Regi√≥n Metropolitana.")
```

```python
# Celda 6: Agrupaci√≥n por Comuna
# Ya tenemos df_simce_rm_raw filtrado solo con RM

# 1. Limpieza de filas sin puntaje
df_proc = df_simce_rm_raw.dropna(subset=['prom_lect4b_rbd', 'prom_mate4b_rbd']).copy()

# 2. ASEGURAR TIPOS ENTEROS (ESTO ES CR√çTICO)
# Forzar a int, es la causa m√°s probable del fallo en el merge de a√±os individuales
df_proc['cod_com_rbd'] = pd.to_numeric(df_proc['cod_com_rbd'], errors='coerce').fillna(0).astype(int)
df_proc['agno'] = df_proc['agno'].astype(int)

# 3. Agrupaci√≥n: Promedio por Comuna y A√±o
df_comunas = df_proc.groupby(['agno', 'cod_com_rbd']).agg({
    'prom_lect4b_rbd': 'mean',
    'prom_mate4b_rbd': 'mean',
    'nom_com_rbd': 'first' 
}).reset_index()

# 4. Renombrar para visualizaci√≥n (Match con Celda 8)
df_comunas.rename(columns={
    'agno': 'A√±o',
    'cod_com_rbd': 'Codigo Comuna',
    'nom_com_rbd': 'Comuna', 
    'prom_lect4b_rbd': 'Avg Reading Score', 
    'prom_mate4b_rbd': 'Avg Math Score'
}, inplace=True)

print(f"Datos agrupados listos: {len(df_comunas)} registros comunales.")
print("A√±os presentes:", sorted(df_comunas['A√±o'].unique()))
```

# Celda que arregla problema

```python
# Celda 6: Agrupaci√≥n con Auto-Reparaci√≥n de C√≥digos (Fix Definitivo 2015/2017)

print("--- INICIANDO PROTOCOLO DE REPARACI√ìN DE C√ìDIGOS ---")

# 1. Normalizaci√≥n de Texto (Para asegurar coincidencia de nombres)
# Quitamos acentos y pasamos a may√∫sculas para que '√ëu√±oa' coincida con 'NUNOA'
def normalize_text(s):
    if pd.isna(s): return ""
    # Reemplazos b√°sicos de acentos
    text = str(s).upper().strip()
    replacements = (("√Å", "A"), ("√â", "E"), ("√ç", "I"), ("√ì", "O"), ("√ö", "U"), ("√ë", "N"))
    for a, b in replacements:
        text = text.replace(a, b)
    return text

# Creamos columna temporal normalizada
df_simce_rm_raw['nom_norm'] = df_simce_rm_raw['nom_com_rbd'].apply(normalize_text)

# 2. Construcci√≥n del Diccionario Maestro
# Usamos los a√±os "buenos" (donde el c√≥digo es un CUT v√°lido de 5 d√≠gitos, ej: 13xxx)
# Filtramos c√≥digos > 10000 para asegurarnos que son formato CUT real
df_good_codes = df_simce_rm_raw[df_simce_rm_raw['cod_com_rbd'] > 10000]

# Creamos un mapa: {NOMBRE_COMUNA_NORMALIZADO : C√ìDIGO_CORRECTO}
# Ej: {'SANTIAGO': 13101, 'PUENTE ALTO': 13201}
master_code_dict = df_good_codes.set_index('nom_norm')['cod_com_rbd'].to_dict()

print(f"‚úÖ Diccionario maestro generado con {len(master_code_dict)} comunas validadas.")

# 3. Aplicaci√≥n de la Reparaci√≥n
def repair_commune_code(row):
    current_code = row['cod_com_rbd']
    # Si el c√≥digo es "corto" (ej: 1, 101) o inv√°lido (< 10000), lo buscamos en el diccionario
    if current_code < 10000:
        # Recuperamos el c√≥digo correcto usando el nombre de la comuna
        repaired_code = master_code_dict.get(row['nom_norm'], current_code)
        return repaired_code
    return current_code

# Sobrescribimos el c√≥digo con la versi√≥n reparada
df_simce_rm_raw['cod_com_rbd_fixed'] = df_simce_rm_raw.apply(repair_commune_code, axis=1)

# Diagn√≥stico de reparaci√≥n
repaired_count = len(df_simce_rm_raw[df_simce_rm_raw['cod_com_rbd'] != df_simce_rm_raw['cod_com_rbd_fixed']])
print(f"üîß Se repararon autom√°ticamente {repaired_count} registros con c√≥digos err√≥neos (2015/2017).")


# 4. Agrupaci√≥n Final (Usando el c√≥digo FIXED)
df_proc = df_simce_rm_raw.dropna(subset=['prom_lect4b_rbd', 'prom_mate4b_rbd'], how='all').copy()

# Agrupamos por A√±o y el C√ìDIGO REPARADO
df_comunas = df_proc.groupby(['agno', 'cod_com_rbd_fixed']).agg({
    'prom_lect4b_rbd': 'mean',
    'prom_mate4b_rbd': 'mean',
    'nom_com_rbd': 'first' # Mantenemos el nombre original para visualizaci√≥n
}).reset_index()

# 5. Renombrado y Tipos (Para el mapa)
df_comunas.rename(columns={
    'agno': 'A√±o',
    'cod_com_rbd_fixed': 'Codigo Comuna',
    'nom_com_rbd': 'Comuna', 
    'prom_lect4b_rbd': 'Avg Reading Score', 
    'prom_mate4b_rbd': 'Avg Math Score'
}, inplace=True)

# Asegurar enteros para el merge
df_comunas['Codigo Comuna'] = df_comunas['Codigo Comuna'].astype(int)

print(f"\nDatos listos para visualizaci√≥n: {len(df_comunas)} registros.")
print("A√±os disponibles:", sorted(df_comunas['A√±o'].unique()))
```

```python
# Celda 7: Visualizaci√≥n Base de la Regi√≥n Metropolitana
# ... (El c√≥digo de b√∫squeda del SHP se mantiene igual)
path_search = os.path.join("Datos", "Comunas", "**", "*.shp")
shapefiles = glob.glob(path_search, recursive=True)

if shapefiles:
    map_path = shapefiles[0]
    print(f"Cargando geometr√≠a desde: {map_path}")
    
    gdf_base = gpd.read_file(map_path)
    
    # 2. Estandarizaci√≥n de columnas para el cruce
    gdf_base.columns = [col.lower() for col in gdf_base.columns]
    
    # Detectar columna de c√≥digo de comuna en el shapefile
    possible_id_cols = ['cod_comuna', 'cut_comuna', 'comuna_id', 'codigo']
    global shp_id_col # Declaramos global para que Celda 8 pueda usarla
    shp_id_col = next((c for c in possible_id_cols if c in gdf_base.columns), None)
    
    if shp_id_col:
        # Forzar tipo entero en el mapa (ESTO ES CR√çTICO)
        gdf_base[shp_id_col] = pd.to_numeric(gdf_base[shp_id_col], errors='coerce').fillna(0).astype(int)
        
        # 3. Filtrado Seguro (Intersection)
        codigos_rm = df_comunas['Codigo Comuna'].unique()
        gdf_rm_map = gdf_base[gdf_base[shp_id_col].isin(codigos_rm)].copy()
        
        # ... (El resto del c√≥digo de visualizaci√≥n de Celda 7)
        if not gdf_rm_map.empty:
             # 4. Visualizaci√≥n limpia
             fig, ax = plt.subplots(figsize=(12, 10))
             gdf_rm_map.plot(ax=ax, color='white', edgecolor='black', linewidth=0.8)
             
             ax.set_title("Divisi√≥n Comunal - Regi√≥n Metropolitana", fontsize=14)
             ax.set_axis_off() 
             plt.show()
             
             print(f"Visualizaci√≥n generada con {len(gdf_rm_map)} comunas.")
        else:
            print("Error: No hay coincidencia de c√≥digos entre el mapa y el DataFrame.")
            print(f"Muestra c√≥digos mapa ({shp_id_col}): {gdf_base[shp_id_col].dropna().unique()[:5]}")
            print(f"Muestra c√≥digos tus datos: {codigos_rm[:5]}")
    else:
        print("Error: No se detect√≥ una columna de c√≥digo de comuna en el shapefile.")
        print(f"Columnas disponibles: {list(gdf_base.columns)}")
else:
    print("Error Cr√≠tico: No se encontr√≥ ning√∫n archivo .shp en la ruta Datos/Comunas.")
```

```python
# Celda 8: Visualizaci√≥n Final - CORRECCI√ìN DE TIPOS FORZADA
# --- 1. CONFIGURACI√ìN Y CORRECCI√ìN DE INGENIER√çA ---
# Aseguramos proyecci√≥n Web para el mapa
gdf_base_web = gdf_rm_map.to_crs(epsg=4326).copy()

# === SOLUCI√ìN DEFINITIVA AL ERROR DE TIPOS ===
# Convertimos la columna ID del mapa de Float (13101.0) a Int (13101)
# Esto alinea el mapa con tus datos de Excel.
gdf_base_web[shp_id_col] = gdf_base_web[shp_id_col].fillna(0).astype(int)

print("‚úÖ Mapa corregido: C√≥digos convertidos a Enteros.")
# ==============================================

# Opciones de A√±os
available_years = sorted(df_comunas['A√±o'].unique())
year_options = ['Promedio Hist√≥rico'] + available_years

# Estado inicial
current_state = {
    'score_type': 'Lenguaje',
    'year': 'Promedio Hist√≥rico'
}

# --- 2. L√ìGICA DE DATOS ---

def get_data_for_plot(year, score_type):
    # A. Filtrado de datos
    if year == 'Promedio Hist√≥rico':
        df_filtered = df_comunas.groupby('Codigo Comuna').agg({
            'Avg Reading Score': 'mean',
            'Avg Math Score': 'mean',
            'Comuna': 'first' 
        }).reset_index()
        title_suffix = "(Promedio 2014-2024)"
    else:
        df_filtered = df_comunas[df_comunas['A√±o'] == year].copy()
        title_suffix = f"({year})"

    # B. Renombrado para Tooltip
    df_filtered = df_filtered.rename(columns={'Comuna': 'Nombre_Comuna_Full'})

    # C. Asegurar que los datos tambi√©n sean INT (Doble seguridad)
    df_filtered['Codigo Comuna'] = df_filtered['Codigo Comuna'].fillna(0).astype(int)

    # D. Configuraci√≥n de Visualizaci√≥n
    if score_type == 'Lenguaje':
        col_target = 'Avg Reading Score'
        base_title = "Simce Lenguaje"
        color_scheme = 'yelloworangered'
    else:
        col_target = 'Avg Math Score'
        base_title = "Simce Matem√°tica"
        color_scheme = 'yellowgreenblue'

    # E. Merge (Ahora seguro: INT con INT)
    gdf_final = gdf_base_web.merge(
        df_filtered,
        left_on=shp_id_col, 
        right_on='Codigo Comuna',
        how='inner'
    )
    
    # F. Helper para Tooltip
    gdf_final['Periodo_View'] = str(year)
    full_title = f"{base_title} {title_suffix}"
    
    return gdf_final, col_target, full_title, color_scheme

# --- 3. RENDERIZADO ---

output_map = widgets.Output()

def render_map():
    year = current_state['year']
    score = current_state['score_type']
    
    try:
        gdf_plot, col_target, title_text, scheme = get_data_for_plot(year, score)
        
        with output_map:
            clear_output(wait=True)
            
            if gdf_plot.empty:
                print(f"‚ö†Ô∏è Alerta: El cruce de datos gener√≥ 0 comunas para el a√±o {year}.")
                print(f"Revisa: Mapa tiene {len(gdf_base_web)} registros, Datos tienen {len(df_comunas[df_comunas['A√±o']==year])} registros.")
            else:
                chart = alt.Chart(gdf_plot).mark_geoshape(
                    stroke='gray', 
                    strokeWidth=0.5
                ).encode(
                    color=alt.Color(
                        col_target, 
                        title="Puntaje",
                        scale=alt.Scale(scheme=scheme)
                    ),
                    tooltip=[
                        alt.Tooltip('Nombre_Comuna_Full', title='Comuna'), 
                        alt.Tooltip(col_target, title='Puntaje', format='.1f'),
                        alt.Tooltip('Periodo_View', title='Periodo')
                    ]
                ).properties(
                    title=title_text,
                    width=600,
                    height=500
                ).project(
                    type='identity', reflectY=True
                )
                display(chart)
                
    except Exception as e:
        with output_map:
            clear_output(wait=True)
            print(f"Error t√©cnico: {e}")

# --- 4. CONTROLES ---

btn_lenguaje = widgets.Button(description='Lenguaje', button_style='warning', icon='book', layout=widgets.Layout(width='180px'))
btn_matematica = widgets.Button(description='Matem√°tica', button_style='info', icon='calculator', layout=widgets.Layout(width='180px'))
dropdown_year = widgets.Dropdown(options=year_options, value='Promedio Hist√≥rico', description='A√±o:', layout=widgets.Layout(width='200px'))

def on_click_lenguaje(b):
    current_state['score_type'] = 'Lenguaje'
    render_map()

def on_click_matematica(b):
    current_state['score_type'] = 'Matematica'
    render_map()

def on_change_year(change):
    if change['type'] == 'change' and change['name'] == 'value':
        current_state['year'] = change['new']
        render_map()

btn_lenguaje.on_click(on_click_lenguaje)
btn_matematica.on_click(on_click_matematica)
dropdown_year.observe(on_change_year)

controls = widgets.HBox([btn_lenguaje, btn_matematica, widgets.Label(" | "), dropdown_year], layout=widgets.Layout(justify_content='center', align_items='center', margin='0 0 15px 0'))

display(controls)
display(output_map)
render_map()
```
